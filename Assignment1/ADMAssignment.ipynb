```python
#Cleaning Customers data
```

```python
df_cust = pd.read_csv('/Users/harooniqbal/Downloads/Customer.csv')
df_cust.info()
```

```python
df_cust
```

```python
#identifying missing values in sales data
plt.figure(figsize=(10, 6))
sns.heatmap(df_cust.isnull(), yticklabels=False, cmap='magma', cbar=False)
```

```python
#dropping customer segment column as it is not required
df_cust.drop('Customer_Segment', inplace=True, axis=1)
df_cust.head()
```

```python
#standardise customer address
df_cust['address_std'] = df_cust['Address'].str.lower()
df_cust['Address'] = df_cust['Address'].str.replace(',', ';', regex = True) # remove comma.
df_cust['address_std'] = df_cust['address_std'].str.strip() # remove leading and trailing whitespace.
df_cust['address_std'] = df_cust['address_std'].str.replace('.', '', regex = True) # remove period.
df_cust['address_std'] = df_cust['address_std'].str.replace('-', '', regex = True) # remove hyphen.
df_cust['address_std'] = df_cust['address_std'].str.replace(',', ';', regex = True) # remove comma.
df_cust['address_std'] = df_cust['address_std'].str.replace('\\,', ';', regex = True) # remove comma.
df_cust['address_std'] = df_cust['address_std'].str.replace('\\bstreet\\b', 'st', regex = True) # replace street with st.
df_cust['address_std'] = df_cust['address_std'].str.replace('\\bSTREET\\b', 'st', regex = True) # replace street with st.
df_cust['address_std'] = df_cust['address_std'].str.replace('\\bapartment\\b', 'apt', regex = True) # replace apartment with apt.
df_cust['address_std'] = df_cust['address_std'].str.replace('\\bAPARTMENT\\b', 'apt', regex = True) # replace apartment with apt.
df_cust['address_std'] = df_cust['address_std'].str.replace('\\bavenue\\b', 'ave', regex = True) # replace apartment with apt.
df_cust['address_std'] = df_cust['address_std'].str.replace('\\bAVENUE\\b', 'ave', regex = True) # replace apartment with apt.

df_cust
```

```python
#correct the spelling mistakes in states names
from nltk.metrics import edit_distance

df_cust['distance_alabama'] = df_cust['State'].map(lambda x: edit_distance(x, 'alabama'))
df_cust['distance_mass'] = df_cust['State'].map(lambda x: edit_distance(x, 'massachusetts'))

msk = df_cust['distance_alabama'] <= 2
df_cust.loc[msk, 'State'] = 'Alabama'
msk = df_cust['distance_mass'] <= 2
df_cust.loc[msk, 'State'] = 'Massachusetts'

df_cust
```
