```python
#creating offensive words list
offensive_words_file = open("/Users/harooniqbal/Desktop/new/NEU_SushmithaJogula/Semester3/ADM/Assignment2/Offensive_Words_List.csv", "r")
csv_reader = csv.reader(offensive_words_file)
offensive_words_list = []
for row in csv_reader:
    #print((row[0]))
    offensive_words_list.append(row[0])

print(offensive_words_list)
```

```python
#method for identifying if the review contains URL or no
def check_url(x):
# Regular expression for url
    re_euq = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?«»“”‘’]))"
    ck_url = re.findall(re_euq, x)
    if ck_url:
        return True
    else:
        return False
```

```python
#method for checking if a review can be published or no
def check_review(review):
    #res = any(ele in review.split(' ') for ele in offensive_words_list)
    for ele in (str(review)).split(' '):
        #print(ele)
        if (len(review)<=100) or (len(review)>800) or (ele in offensive_words_list) or (check_url(review) is True):
        #if (check_url(review) is True) and 1<5:
            #print("here")
            return 'Fail'
    return 'Publish'    
        
```

```python
#checking the status of review
review_status = []
for (review,label) in zip(reviews_df.loc[:,'Reviews'], reviews_df.loc[:,'R/ F']): 
    if label == 'R':
        if (check_review(review)=='Publish'):
            review_status.append('Publish')
        else:
            review_status.append('Fail')
    else:
        review_status.append('Fail')    
reviews_df['ReviewStatus'] = review_status
```
